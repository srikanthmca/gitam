#  1. Project on IPL Data set

- Display the count of matches based on the year wise

- Plot a bar graph  no. of matches won by each team in the year  2011

- Print  ManOfMatch count of each player in  Hyderabad stadium

```python
# 1. Project on IPL Data set
import pandas as pd 
filePath='DataFile/Match.csv'
def readCsvData (filePath):
    return pd.read_csv(filePath)
df=readCsvData(filePath)
print(df)
```

```python
#1.1 Display the count of matches based on the year wise
df["Season_Year"].value_counts()
```

```python
#1.2 Plot a bar graph  no. of matches won by each team in the year  2011
import matplotlib.pyplot as plt
import seaborn as sns
rows_2011 = df[df.Season_Year == 2011]
sns.countplot(y = rows_2011['match_winner'])
plt.xlabel('winner frequency')
plt.title('win data for the year 2011')
plt.show()
```

```python
#1.3 Print  ManOfMatch count of each player in  Hyderabad stadium
Hyderabad=df[df.City_Name==
'Hyderabad']
motm=pd.crosstab(index=Hyderabad["ManOfMach"],columns='count',margins=False)
print(motm)
```

# 2. Project on Appstore Data-set

- Calculate the average rating of Gaming and Non-Gaming apps 
- Draw Pie plot based on labels as  below Price category and values as count of each category apps labels:
    - Price =0  - Free apps
    - Price <20 - Affordable apps
    - Price<50 - Expensive apps
    - Price >50 - Very expensive

```python
# 2.Project on Appstore Data-set
import pandas as pd
filepath ='DataFile/AppleStore.csv'
def readCsvData(filepath):
    return pd.read_csv(filepath)
df = readCsvData(filepath)
print(df)
```

```python
# 2.1 Calculate the average rating of Gaming and Non-Gaming apps
t= df[['prime_genre','user_rating']]
g= t[t.prime_genre =='Games']
ng=t[t.prime_genre !='Games']
print("Mean {0}".format(g.mean()))
print("Mean non games{0}".format(ng.mean()))
```

```python
# 2.2 Draw Pie plot based on labels as below Price category and values as count of each category apps labels:
# Price =0 - Free apps
# Price <20 - Affordable apps
# Price<50 - Expensive apps
# Price >50 - Very expensiv
movies=['Free apps','AffordableApps','Expensive apps','Very Expensive']
rating1=df[df.price==0]
rating2=df[(df.price>0) & (df.price<20)]
rating3=df[(df.price>20) & (df.price<50)]
rating4=df[(df.price>50)]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea= rating3.count()[0]
ve= rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90)
plt.show()
```
